### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Añadir identificador de tienda a cada DataFrame
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Calcular el ingreso total de cada tienda sumando la columna Precio
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostrar los resultados
print("Ingreso total de Tienda 1:", ingreso_tienda1)
print("Ingreso total de Tienda 2:", ingreso_tienda2)
print("Ingreso total de Tienda 3:", ingreso_tienda3)
print("Ingreso total de Tienda 4:", ingreso_tienda4)
```

```python
#Crear el gráfico de barras con matplotlib (ajustado)
from matplotlib import pyplot as plt
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]  # Datos reales
ingresos_millones = [i / 1000000 for i in ingresos]  # Dividir entre un millón para mejor escala
plt.bar(tiendas, ingresos_millones, color=['red', 'green', 'blue', 'yellow'])  # Colores simples
plt.title("Ingresos de las Tiendas (en millones)")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos (millones)")

# Añadir los valores encima de las barras
for i, v in enumerate(ingresos_millones):
    plt.text(i, v + 10, str(round(v, 2)), ha='center')  # Muestra el valor redondeado

plt.show()
```

# 2. Ventas por categoría

```python

# Calcular la cantidad de productos vendidos por categoría en cada tienda
ventas_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Mostrar los resultados para cada tienda
print("Cantidad de productos vendidos por categoría en Tienda 1:")
print(ventas_categoria_tienda1)
print("\nCantidad de productos vendidos por categoría en Tienda 2:")
print(ventas_categoria_tienda2)
print("\nCantidad de productos vendidos por categoría en Tienda 3:")
print(ventas_categoria_tienda3)
print("\nCantidad de productos vendidos por categoría en Tienda 4:")
print(ventas_categoria_tienda4)
```

```python
# Crear gráficos de pastel para cada tienda (top 3 categorías)
from matplotlib import pyplot as plt

# Tienda 1
categorias_t1 = ventas_categoria_tienda1.index.tolist()[:8]
conteos_t1 = ventas_categoria_tienda1.values.tolist()[:8]
plt.pie(conteos_t1, labels=categorias_t1, colors=['red', 'green', 'blue'], autopct='%1.1f%%')  # Porcentajes
plt.title("Categorías más vendidas en Tienda 1")
plt.show()

# Tienda 2
categorias_t2 = ventas_categoria_tienda2.index.tolist()[:8]
conteos_t2 = ventas_categoria_tienda2.values.tolist()[:8]
plt.pie(conteos_t2, labels=categorias_t2, colors=['red', 'green', 'blue'], autopct='%1.1f%%')
plt.title("Categorías más vendidas en Tienda 2")
plt.show()

# Tienda 3
categorias_t3 = ventas_categoria_tienda3.index.tolist()[:8]
conteos_t3 = ventas_categoria_tienda3.values.tolist()[:8]
plt.pie(conteos_t3, labels=categorias_t3, colors=['red', 'green', 'blue'], autopct='%1.1f%%')
plt.title("Categorías más vendidas en Tienda 3")
plt.show()

# Tienda 4
categorias_t4 = ventas_categoria_tienda4.index.tolist()[:8]
conteos_t4 = ventas_categoria_tienda4.values.tolist()[:8]
plt.pie(conteos_t4, labels=categorias_t4, colors=['red', 'green', 'blue'], autopct='%1.1f%%')
plt.title("Categorías más vendidas en Tienda 4")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio de cada tienda
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

# Mostrar los resultados
print("Calificación promedio de Tienda 1:", calificacion_promedio_tienda1)
print("Calificación promedio de Tienda 2:", calificacion_promedio_tienda2)
print("Calificación promedio de Tienda 3:", calificacion_promedio_tienda3)
print("Calificación promedio de Tienda 4:", calificacion_promedio_tienda4)



```

```python
# Crear el gráfico de líneas con matplotlib
from matplotlib import pyplot as plt
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]
plt.plot(tiendas, calificaciones, color='blue', marker='o')  # Línea con puntos
plt.title("Calificaciones Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación")
plt.ylim(0, 5)  # Rango de 0 a 5
for i, v in enumerate(calificaciones):
    plt.text(i, v + 0.1, str(round(v, 4)), ha='center')  # Muestra el valor redondeado
plt.show()
```

# 4. Productos más y menos vendidos

```python

# Calcular la frecuencia de productos vendidos en cada tienda
productos_tienda1 = tienda['Producto'].value_counts()
productos_tienda2 = tienda2['Producto'].value_counts()
productos_tienda3 = tienda3['Producto'].value_counts()
productos_tienda4 = tienda4['Producto'].value_counts()

# Identificar el producto más vendido y menos vendido en cada tienda
# Más vendido: el de mayor conteo
# Menos vendido: el de menor conteo (excluyendo 1 si hay muchos productos únicos)
mas_vendido_tienda1 = productos_tienda1.idxmax()
menos_vendido_tienda1 = productos_tienda1.idxmin() if productos_tienda1.min() > 1 else productos_tienda1[productos_tienda1 > 1].idxmin()
mas_vendido_tienda2 = productos_tienda2.idxmax()
menos_vendido_tienda2 = productos_tienda2.idxmin() if productos_tienda2.min() > 1 else productos_tienda2[productos_tienda2 > 1].idxmin()
mas_vendido_tienda3 = productos_tienda3.idxmax()
menos_vendido_tienda3 = productos_tienda3.idxmin() if productos_tienda3.min() > 1 else productos_tienda3[productos_tienda3 > 1].idxmin()
mas_vendido_tienda4 = productos_tienda4.idxmax()
menos_vendido_tienda4 = productos_tienda4.idxmin() if productos_tienda4.min() > 1 else productos_tienda4[productos_tienda4 > 1].idxmin()

# Contar las ventas de cada producto identificado
ventas_mas_tienda1 = productos_tienda1.max()
ventas_menos_tienda1 = productos_tienda1[productos_tienda1 > 1].min() if productos_tienda1.min() == 1 else productos_tienda1.min()
ventas_mas_tienda2 = productos_tienda2.max()
ventas_menos_tienda2 = productos_tienda2[productos_tienda2 > 1].min() if productos_tienda2.min() == 1 else productos_tienda2.min()
ventas_mas_tienda3 = productos_tienda3.max()
ventas_menos_tienda3 = productos_tienda3[productos_tienda3 > 1].min() if productos_tienda3.min() == 1 else productos_tienda3.min()
ventas_mas_tienda4 = productos_tienda4.max()
ventas_menos_tienda4 = productos_tienda4[productos_tienda4 > 1].min() if productos_tienda4.min() == 1 else productos_tienda4.min()

# Mostrar los resultados
print("Tienda 1 - Producto más vendido:", mas_vendido_tienda1, "con", ventas_mas_tienda1, "ventas")
print("Tienda 1 - Producto menos vendido:", menos_vendido_tienda1, "con", ventas_menos_tienda1, "ventas")
print("\nTienda 2 - Producto más vendido:", mas_vendido_tienda2, "con", ventas_mas_tienda2, "ventas")
print("Tienda 2 - Producto menos vendido:", menos_vendido_tienda2, "con", ventas_menos_tienda2, "ventas")
print("\nTienda 3 - Producto más vendido:", mas_vendido_tienda3, "con", ventas_mas_tienda3, "ventas")
print("Tienda 3 - Producto menos vendido:", menos_vendido_tienda3, "con", ventas_menos_tienda3, "ventas")
print("\nTienda 4 - Producto más vendido:", mas_vendido_tienda4, "con", ventas_mas_tienda4, "ventas")
print("Tienda 4 - Producto menos vendido:", menos_vendido_tienda4, "con", ventas_menos_tienda4, "ventas")
```

```python
from matplotlib import pyplot as plt

# Tienda 1
plt.bar([mas_vendido_tienda1, menos_vendido_tienda1], [ventas_mas_tienda1, ventas_menos_tienda1], color=['green', 'red'])
plt.title("Productos Más y Menos Vendidos en Tienda 1")
plt.ylabel("Número de Ventas")
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas si los nombres son largos
for i, v in enumerate([ventas_mas_tienda1, ventas_menos_tienda1]):
    plt.text(i, v + 1, str(v), ha='center')
plt.show()

# Tienda 2
plt.bar([mas_vendido_tienda2, menos_vendido_tienda2], [ventas_mas_tienda2, ventas_menos_tienda2], color=['green', 'red'])
plt.title("Productos Más y Menos Vendidos en Tienda 2")
plt.ylabel("Número de Ventas")
plt.xticks(rotation=45, ha='right')
for i, v in enumerate([ventas_mas_tienda2, ventas_menos_tienda2]):
    plt.text(i, v + 1, str(v), ha='center')
plt.show()

# Tienda 3
plt.bar([mas_vendido_tienda3, menos_vendido_tienda3], [ventas_mas_tienda3, ventas_menos_tienda3], color=['green', 'red'])
plt.title("Productos Más y Menos Vendidos en Tienda 3")
plt.ylabel("Número de Ventas")
plt.xticks(rotation=45, ha='right')
for i, v in enumerate([ventas_mas_tienda3, ventas_menos_tienda3]):
    plt.text(i, v + 1, str(v), ha='center')
plt.show()

# Tienda 4
plt.bar([mas_vendido_tienda4, menos_vendido_tienda4], [ventas_mas_tienda4, ventas_menos_tienda4], color=['green', 'red'])
plt.title("Productos Más y Menos Vendidos en Tienda 4")
plt.ylabel("Número de Ventas")
plt.xticks(rotation=45, ha='right')
for i, v in enumerate([ventas_mas_tienda4, ventas_menos_tienda4]):
    plt.text(i, v + 1, str(v), ha='center')
plt.show()
```

## Informe Final — Recomendación de Venta

Luego de analizar las métricas clave de las cuatro tiendas, se llega a una conclusión clara sobre cuál de ellas representa la menor rentabilidad.

### Ingresos Totales
Tienda 1 tuvo el mejor desempeño financiero, superando los \$1,150 millones en ventas.  
Tienda 4, por otro lado, reportó el ingreso más bajo: \$1,038 millones.

### Categorías de Productos
La Tienda 4 mostró fuerza en muebles y juguetes, mientras que libros e instrumentos musicales siguen siendo las categorías de menor movimiento en todas las tiendas.

### Calificaciones de Clientes
La mejor evaluación la obtuvo Tienda 2 (4.09), seguida por Tienda 3 (4.05).  
Tienda 4 recibió un promedio de 4.00, lo cual es aceptable pero no destaca.

### Productos Más y Menos Vendidos
Aunque Tienda 4 tuvo buenos resultados con televisores, también registró productos de baja rotación como una taza de cerámica. Esto refleja una demanda desigual.

### Costo de Envío
Tienda 4 sobresale en este punto con el menor costo de envío promedio. Sin embargo, este aspecto no compensa sus debilidades en ingresos y ventas.

### Conclusión

Considerando todos los factores — ventas, popularidad de productos, evaluación del cliente y logística — se recomienda vender la Tienda 4. Es la que muestra menor rentabilidad general, y liberar esa operación permitirá invertir en opciones más estratégicas que aporten mayor valor a futuro.
